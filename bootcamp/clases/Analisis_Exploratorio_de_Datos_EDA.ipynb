**Gráfico de Líneas**

Un gráfico de líneas se utiliza para mostrar datos a lo largo del tiempo o en un orden secuencial.

 

**Gráfico de una Variable**

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear el gráfico de líneas
plt.plot(x, y)
plt.title("Gráfico de Líneas de una Variable")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

**Gráfico de Dos Variables**

```python
# Datos de ejemplo
y2 = [1, 4, 6, 8, 10]

# Crear el gráfico de líneas con dos variables
plt.plot(x, y, label='Variable 1')
plt.plot(x, y2, label='Variable 2')
plt.title("Gráfico de Líneas de Dos Variables")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.show()
```

**Personalización de Gráficos de Líneas**

**Tipos de Línea y Estilos**

```python
# Crear el gráfico de líneas con estilo personalizado
plt.plot(x, y, linestyle='--', color='r', marker='o', label='Variable 1')
plt.plot(x, y2, linestyle='-', color='b', marker='x', label='Variable 2')
plt.title("Gráfico de Líneas con Estilos Personalizados")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
plt.show()
```

Matplotlib permite personalizar el estilo de las líneas utilizando diferentes parámetros en la función plot.

**linestyle:** Tipo de línea (‘-‘, ‘–‘, ‘-.’, ‘:’)

**color:** Color de la línea (‘r’, ‘g’, ‘b’, etc. o valores hexadecimales como ‘#ff0000’)

**marker:** Tipo de marcador (‘o’, ‘x’, ‘s’, ‘D’, etc.)

**grid:** Añade una cuadrícula al gráfico

**Ejemplo Completo de Personalización**

```python
# Crear el gráfico de líneas completamente personalizado
plt.plot(x, y, linestyle='--', color='r', marker='o', linewidth=2, markersize=10, label='Variable 1')
plt.plot(x, y2, linestyle='-', color='b', marker='x', linewidth=2, markersize=10, label='Variable 2')
plt.title("Gráfico de Líneas Completamente Personalizado", fontsize=14, fontweight='bold')
plt.xlabel("Eje X", fontsize=12)
plt.ylabel("Eje Y", fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', linewidth=0.5)
plt.show()
```

**Gráfico de Barras**

Un gráfico de barras se utiliza para comparar diferentes categorías o grupos.

**Gráfico de Barras Simples**

```python
categorias = ['A', 'B', 'C', 'D']
valores = [4, 7, 1, 8]

# Crear el gráfico de barras
plt.bar(categorias, valores)
plt.title("Gráfico de Barras")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

**Gráfico de Barras Agrupadas**

```python
import numpy as np

# Datos de ejemplo
valores2 = [5, 6, 2, 7]
width = 0.4
ind = np.arange(len(categorias))

# Crear el gráfico de barras agrupadas
plt.bar(ind - width/2, valores, width, label='Grupo 1')
plt.bar(ind + width/2, valores2, width, label='Grupo 2')
plt.title("Gráfico de Barras Agrupadas")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.xticks(ind, categorias)
plt.legend()
plt.show()
```

**Gráfico de Dispersión**

Un gráfico de dispersión se utiliza para mostrar la relación entre dos variables.

**Gráfico de Dispersión Simple**


```python
# Crear el gráfico de dispersión
plt.scatter(x, y)
plt.title("Gráfico de Dispersión")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
# Crear el gráfico de dispersión con personalización
plt.scatter(x, y, color='r', marker='o', s=100, label='Datos 1')
plt.scatter(x, y2, color='b', marker='x', s=100, label='Datos 2')
plt.title("Gráfico de Dispersión Personalizado")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.show()
```
