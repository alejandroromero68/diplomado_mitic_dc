## **Cuadriculas, leyendas y otros elementos de Visualización**
 
En esta sección, aprenderás cómo añadir cuadrículas y leyendas a tus gráficos para hacerlos más claros y fáciles de interpretar. También exploraremos otros elementos útiles que no hemos cubierto hasta ahora, como anotaciones y flechas. Estas herramientas te ayudarán a crear visualizaciones más informativas y profesionales.
 
**Cuadrículas en Matplotlib**
 
Las cuadrículas son líneas horizontales y verticales que se añaden a un gráfico para ayudar a visualizar la relación entre los puntos de datos y las escalas de los ejes.
 
**Añadir Cuadrículas**

Para añadir una cuadrícula a tu gráfico, utiliza el método grid():

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico
plt.plot(x, y)
plt.title('Gráfico de Seno con Cuadrícula')
plt.xlabel('X')
plt.ylabel('Y')

# Añadir cuadrícula
plt.grid(True)
```

**Personalización de Cuadrículas**
 
Puedes personalizar el aspecto de la cuadrícula utilizando parámetros adicionales:

```python
# Añadir cuadrícula personalizada
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')
```

**which:** ‘both’ para líneas mayores y menores, ‘major’ para solo líneas mayores, ‘minor’ para solo líneas menores.

**linestyle:** Estilo de la línea (por ejemplo, ‘–‘ para línea discontinua).

**linewidth:** Grosor de la línea.

**color:** Color de la línea.

**Leyendas en Matplotlib**
 
Las leyendas explican el significado de diferentes elementos en un gráfico, como las líneas de diferentes colores o estilos.
 
**Añadir Leyendas**
 
Para añadir una leyenda a tu gráfico, utiliza el método legend(). Primero debes etiquetar las líneas o puntos utilizando el parámetro label en las funciones de trazado.

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear un gráfico con leyendas
plt.plot(x, y1, label='Seno')
plt.plot(x, y2, label='Coseno')
plt.title('Gráfico de Seno y Coseno')
plt.xlabel('X')
plt.ylabel('Y')

# Añadir leyenda
plt.legend()

plt.show()
```

**Personalización de Leyendas**

**Anotaciones en Matplotlib**
 
Las anotaciones son textos o símbolos que se añaden a un gráfico para resaltar puntos específicos.
 
**Añadir Anotaciones**
 
Para añadir anotaciones, utiliza el método annotate():

**loc:** Ubicación de la leyenda (por ejemplo, ‘upper right’, ‘lower left’, ‘best’).

**fontsize:** Tamaño del texto de la leyenda.

**shadow:** Añadir sombra a la leyenda.

**title:** Añadir un título a la leyenda.

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico con anotaciones
plt.plot(x, y)
plt.title('Gráfico de Seno con Anotaciones')
plt.xlabel('X')
plt.ylabel('Y')

# Añadir anotación
plt.annotate('Punto máximo', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

**xy:** Coordenadas del punto que se anotará.
**xytext:** Coordenadas donde se ubicará el texto de la anotación.
**arrowprops:** Propiedades de la flecha (por ejemplo, color y tamaño).
 
 
**Flechas en Matplotlib**
 
Las flechas pueden ayudar a señalar puntos específicos en un gráfico.
 
 
**Añadir Flechas**
  
Para añadir flechas, utiliza el método arrow():

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico con flechas
plt.plot(x, y)
plt.title('Gráfico de Seno con Flechas')
plt.xlabel('X')
plt.ylabel('Y')

# Añadir flecha
plt.arrow(np.pi/2, 0.5, 0, 0.4, head_width=0.2, head_length=0.1, fc='green', ec='green')

plt.show()
```

**head_width:** Ancho de la cabeza de la flecha.

**head_length:** Longitud de la cabeza de la flecha.

**fc:** Color de relleno de la flecha.

**ec:** Color del borde de la flecha.
 
 
**Ejemplo Completo**

A continuación, un ejemplo completo que combina cuadrículas, leyendas, anotaciones y flechas para crear un gráfico informativo y atractivo.

```python
import matplotlib.pyplot as plt
import numpy as np

# Aplicar un estilo predefinido
#plt.style.use('seaborn-darkgrid')
plt.style.use('ggplot')  # Un estilo predefinido de Matplotlib


# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear una figura
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de Seno
ax.plot(x, y1, label='Seno', color='b', marker='o', linestyle='-', markersize=5)

# Gráfico de Coseno
ax.plot(x, y2, label='Coseno', color='r', marker='^', linestyle='--', markersize=5)

# Añadir cuadrícula personalizada
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Añadir título y etiquetas
ax.set_title('Gráficos de Seno y Coseno con Elementos Personalizados', fontsize=20)
ax.set_xlabel('Eje X', fontsize=15)
ax.set_ylabel('Eje Y', fontsize=15)

# Añadir leyenda personalizada
ax.legend(loc='upper right', fontsize='large', shadow=True, title='Funciones')

# Añadir anotación
ax.annotate('Punto máximo del Seno', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir flecha
ax.arrow(np.pi/2, 0.5, 0, 0.4, head_width=0.2, head_length=0.1, fc='green', ec='green')

# Mostrar el gráfico
plt.show()
```
