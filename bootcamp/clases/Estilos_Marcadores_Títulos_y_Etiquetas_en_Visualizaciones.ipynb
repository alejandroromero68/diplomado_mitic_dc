## **Estilos, Marcadores, Títulos y Etiquetas en Visualizaciones**

En esta sección del curso, aprenderás cómo personalizar tus gráficos para hacerlos más informativos y visualmente atractivos. Nos centraremos en la aplicación de estilos, el uso de marcadores, y la configuración de títulos y etiquetas en gráficos creados con Matplotlib y Seaborn.
 
**Estilos en Matplotlib**

Matplotlib ofrece una variedad de estilos predefinidos que se pueden aplicar a tus gráficos para cambiar su apariencia. Estos estilos pueden ayudar a que tus gráficos se vean más profesionales y consistentes con la estética de tu presentación o documento.
 
**Aplicación de Estilos**
 
Para aplicar un estilo, utilizamos plt.style.use(). Aquí hay un ejemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

# Aplicar un estilo predefinido
plt.style.use('ggplot')

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico
plt.plot(x, y)
plt.title('Gráfico de Seno con Estilo ggplot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

**Marcadores en Gráficos**
 
Los marcadores son símbolos que se utilizan para resaltar puntos de datos individuales en un gráfico. Pueden ayudar a destacar datos importantes y hacer que los gráficos sean más fáciles de interpretar.
 
**Tipos de Marcadores**
 
Matplotlib ofrece una variedad de marcadores que se pueden aplicar a los puntos de datos. Aquí hay algunos ejemplos:
‘.’: Punto
‘,’: Pixel
‘o’: Círculo
‘v’: Triángulo hacia abajo
‘^’: Triángulo hacia arriba
‘s’: Cuadrado
‘p’: Pentágono
 
**Uso de Marcadores**
 
Aquí hay un ejemplo de cómo utilizar marcadores en un gráfico de dispersión:
 

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico con marcadores
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Gráfico de Seno con Marcadores')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

**Títulos y Etiquetas**
 
Los títulos y etiquetas son esenciales para hacer que los gráficos sean comprensibles. Los títulos proporcionan un resumen del contenido del gráfico, mientras que las etiquetas de los ejes describen las variables que se representan.
 
**Añadir Títulos**

Utilizamos plt.title() para añadir un título a un gráfico. Podemos personalizar el tamaño, color y estilo del título.

```python
# Crear un gráfico con un título
plt.plot(x, y)
plt.title('Gráfico de Seno', fontsize=20, color='r', loc='center')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

**fontsize:** Tamaño del texto del título. 

**color:** Color del texto del título. 

**loc:** Posición del título (‘left’, ‘center’, ‘right’). 

 
**Añadir Etiquetas a los Ejes**

Utilizamos plt.xlabel() y plt.ylabel() para añadir etiquetas a los ejes X e Y, respectivamente.

```python
# Crear un gráfico con etiquetas en los ejes
plt.plot(x, y)
plt.xlabel('Eje X', fontsize=15)
plt.ylabel('Eje Y', fontsize=15)
plt.title('Gráfico de Seno con Etiquetas')
plt.show()
```

**Personalización Avanzada**

Podemos combinar varios elementos para crear gráficos altamente personalizados y profesionales.

**Ejemplo Completo:** Aquí hay un ejemplo completo que combina estilos, marcadores, títulos y etiquetas:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Aplicar un estilo predefinido
#plt.style.use('seaborn-darkgrid')
#plt.style.use('ggplot')  # Otro estilo similar
# Estilos disponibles en Matplotlib
plt.style.use('ggplot')  # Un estilo predefinido de Matplotlib

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear una figura con subplots
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de Seno
ax.plot(x, y1, label='Seno', color='b', marker='o', linestyle='-', markersize=5)

# Gráfico de Coseno
ax.plot(x, y2, label='Coseno', color='r', marker='^', linestyle='--', markersize=5)

# Añadir título y etiquetas
ax.set_title('Gráficos de Seno y Coseno', fontsize=20)
ax.set_xlabel('Eje X', fontsize=15)
ax.set_ylabel('Eje Y', fontsize=15)

# Añadir leyenda
ax.legend(loc='upper right')

# Mostrar el gráfico
plt.show()
```
